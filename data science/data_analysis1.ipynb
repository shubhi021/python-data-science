## selecting rows and columns
- rows (slicing)
   - 'loc' - get us data using named indexes
   - 'iloc' - get us data using numeric indexes
- columns (dictionary style)
    - df[single] - requires name of column
    - df[multuple] - requires list of columns(multiple is list of named columns) 

```python
#single row(record)
weather.iloc[0]
```

```python
#for last record
weather.iloc[-1]
```

```python
#record from 1to 20
weather.iloc[:20]
```

```python
#for record between 25 to 50
weather.iloc[25:50]
```

```python
# record 1to 15 with first 5 columns
weather.iloc[1:16,:5]
```

```python
#15 records with last 3 columns
weather.iloc[:15,-3:]
```

```python
#15 records with last 3 columns
weather.iloc[:15,[-3,-2,-1]]#gives sameas above
```

```python
weather['Date']#selecting a single column(dictionary style) -> Series 
```

```python
weather.AvgTemp #same as weather['AvgTemp](object style) -> series same as above
```

```python
cols =[ 'Date','MaxTemp','MinTemp']
weather[cols] # multiple columns (dictionary style)--> Data Frames
```

```python
weather[['Precipitation','State','City','Date']]#its shorter version and is same as above
```
